# Filter league to run and data manipulation

```python

df_champions = pd.read_excel('/drive/My Drive/Flashscore Leagues/All leagues/'+league_name + '.xlsx')
df_champions.head()
```

```python
df_champions['class_label'] = 'NA'

for i in range(0, df_champions.shape[0]):
    if df_champions['home_score'].iloc[i] < df_champions['away_score'].iloc[i]:
        df_champions['class_label'].iloc[i] = 'away'
    elif df_champions['home_score'].iloc[i] == df_champions['away_score'].iloc[i]:
        df_champions['class_label'].iloc[i] = 'draw'
    elif df_champions['home_score'].iloc[i] > df_champions['away_score'].iloc[i]:
        df_champions['class_label'].iloc[i] = 'home'
    else:
        df_champions['class_label'].iloc[i] = 'null'
        

```

```python
import numpy as np

for i in range(0, df_champions.shape[0]):
    df_champions['Ball_Possession_home'].iloc[i] = np.float64(str(df_champions['Ball_Possession_home'].iloc[i]).rstrip("%"))/100
    df_champions['Ball_Possession_away'].iloc[i] = np.float64(str(df_champions['Ball_Possession_away'].iloc[i]).rstrip("%"))/100
    
    df_champions['Ball_Possession_home_1st_half'].iloc[i] = np.float64(str(df_champions['Ball_Possession_home_1st_half'].iloc[i]).rstrip("%"))/100
    df_champions['Ball_Possession_away_1st_half'].iloc[i] = np.float64(str(df_champions['Ball_Possession_away_1st_half'].iloc[i]).rstrip("%"))/100
    
    df_champions['Ball_Possession_home_2nd_half'].iloc[i] = np.float64(str(df_champions['Ball_Possession_home_2nd_half'].iloc[i]).rstrip("%"))/100
    df_champions['Ball_Possession_away_2nd_half'].iloc[i] = np.float64(str(df_champions['Ball_Possession_away_2nd_half'].iloc[i]).rstrip("%"))/100
df_champions.head(10)
```

```python
df_champions['Ball_Possession_home'] = df_champions['Ball_Possession_home'].astype(float)
df_champions['Ball_Possession_away'] = df_champions['Ball_Possession_away'].astype(float)

df_champions['Ball_Possession_home_1st_half'] = df_champions['Ball_Possession_home_1st_half'].astype(float)
df_champions['Ball_Possession_away_1st_half'] = df_champions['Ball_Possession_away_1st_half'].astype(float)

df_champions['Ball_Possession_home_2nd_half'] = df_champions['Ball_Possession_home_2nd_half'].astype(float)
df_champions['Ball_Possession_away_2nd_half'] = df_champions['Ball_Possession_away_2nd_half'].astype(float)

df_champions.dtypes
```

```python
df_champions['date_formatted'] = pd.to_datetime(df_champions['date'], dayfirst = True)
df_champions['week'] = df_champions['date_formatted'].dt.isocalendar().week
df_champions['year'] = df_champions['date_formatted'].dt.isocalendar().year
df_champions.head()
```

```python
import math
def round_up(n, decimals=0):
    multiplier = 10 ** decimals
    return math.ceil(n * multiplier) / multiplier
```

```python
df_champions_seasons_week = df_champions.groupby(['season']).date_formatted.agg(['min']).reset_index()

df_champions = df_champions.merge(df_champions_seasons_week, on = ['season'], how = 'left')

df_champions['days'] = df_champions['date_formatted'] - df_champions['min']
df_champions['week'] = df_champions['days']/7
df_champions['weeks'] = df_champions['week'] / pd.to_timedelta(1, unit='D')


df_champions['weeks_rounded'] = 0
for i in range(0, df_champions.shape[0]):
  df_champions['weeks_rounded'].iloc[i] = round_up(df_champions['weeks'].iloc[i], decimals = 0)
df_champions = df_champions.drop(['week', 'year', 'min', 'days', 'weeks'], axis = 1)
df_champions = df_champions.rename(columns={"weeks_rounded": "week"})
```

```python
df_champions.groupby(['season']).date_formatted.agg(['min', 'max']).reset_index()
```

```python
df_champions['match_points_home'] = 0
df_champions['match_points_away'] = 0

for i in range(0, df_champions.shape[0]):
    if df_champions['class_label'].iloc[i] == 'home':
        df_champions['match_points_home'].iloc[i] = 3
        df_champions['match_points_away'].iloc[i] = 0
    elif df_champions['class_label'].iloc[i] == 'away':
        df_champions['match_points_home'].iloc[i] = 0
        df_champions['match_points_away'].iloc[i] = 3
    elif df_champions['class_label'].iloc[i] == 'draw':
        df_champions['match_points_home'].iloc[i] = 1
        df_champions['match_points_away'].iloc[i] = 1
    
df_champions.head()     
```

```python
f = df_champions.groupby(['season','home_team', 'week', 'date_formatted'])['match_points_home'].sum().reset_index()
g = df_champions.groupby(['season','away_team', 'week', 'date_formatted'])['match_points_away'].sum().reset_index()
```

```python
f.columns = ['season', 'home_team', 'week','date_formatted', 'weekly_points_home']
g.columns = ['season', 'away_team','week','date_formatted', 'weekly_points_away']
```

```python
f = f.sort_values(by=['season', 'home_team','week','date_formatted'], ascending = True)
g = g.sort_values(by=['season','away_team','week','date_formatted'], ascending = True)
```

```python
df_final = pd.merge(df_champions, f, how = 'left', on = ['home_team', 'season', 'week','date_formatted'])
df_final = pd.merge(df_final, g, how = 'left', on = ['away_team', 'season', 'week','date_formatted'])
df_final.head(2)
```

```python
f = df_final.groupby(['season','home_team', 'week', 'date_formatted', 'match_points_home'])['Ball_Possession_home'].sum().reset_index()
g = df_final.groupby(['season','away_team', 'week', 'date_formatted', 'match_points_away'])['Ball_Possession_away'].sum().reset_index()

f.columns = ['season', 'home_team', 'week','date_formatted', 'match_points_home', 'random']
g.columns = ['season', 'away_team', 'week','date_formatted', 'match_points_away', 'random']

f = f.drop('random', axis = 1)
g = g.drop('random', axis = 1)


f['weekly_points_cumsum_home'] = 0
cumsumhome = f.groupby(['season', 'home_team'])['match_points_home'].cumsum().reset_index()

for i in range(0, f.shape[0]):
    f['weekly_points_cumsum_home'].iloc[i] = cumsumhome['match_points_home'].iloc[i]

    
g['weekly_points_cumsum_away'] = 0
cumsumaway = g.groupby(['season', 'away_team'])['match_points_away'].cumsum().reset_index()

for i in range(0, g.shape[0]):
    g['weekly_points_cumsum_away'].iloc[i] = cumsumaway['match_points_away'].iloc[i]
    

f['cumsum_till_previous_week_home'] = f['weekly_points_cumsum_home'] - f['match_points_home']
g['cumsum_till_previous_week_away'] = g['weekly_points_cumsum_away'] - g['match_points_away']


### qcing for chelsea
fff = f[f['home_team'] == 'Chelsea'] 
fff = fff[fff['season'] == 'Premier League 2016/2017'].sort_values('date_formatted', ascending = False)
fff
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import BernoulliRBM

def run_models(X_train_scaled, y_train):
  gb = GradientBoostingClassifier()
  gb.fit(X_train_scaled, y_train)


  rf = RandomForestClassifier()
  rf.fit(X_train_scaled, y_train)


  dt = DecisionTreeClassifier()
  dt.fit(X_train_scaled, y_train)

  lr = LogisticRegression()
  lr.fit(X_train_scaled, y_train)

  mlp = MLPClassifier(hidden_layer_sizes = 100, activation = 'identity', solver = 'lbfgs',learning_rate = 'adaptive', learning_rate_init = 0.0001, alpha = 0.00001)
  mlp.fit(X_train_scaled, y_train)

  return gb, rf, dt, lr, mlp
```

```python
import warnings
warnings.filterwarnings('ignore')


df_abc = pd.DataFrame()
df_abc.index = range(2, last_n_matches+1)
df_abc['dt'] = 0
df_abc['rf'] = 0
df_abc['gbm'] = 0
df_abc['lr'] = 0
df_abc['mlp'] = 0

df_abc['dt_profit'] = 0
df_abc['rf_profit'] = 0
df_abc['gbm_profit'] = 0
df_abc['lr_profit'] = 0
df_abc['mlp_profit'] = 0

df_abc['dt_profit_prob_diff_filter'] = 0
df_abc['rf_profit_prob_diff_filter'] = 0
df_abc['gbm_profit_prob_diff_filter'] = 0
df_abc['lr_profit_prob_diff_filter'] = 0
df_abc['mlp_profit_prob_diff_filter'] = 0

df_abc['dt_profit_std_dev_filter'] = 0
df_abc['rf_profit_std_dev_filter'] = 0
df_abc['gbm_profit_std_dev_filter'] = 0
df_abc['lr_profit_std_dev_filter'] = 0
df_abc['mlp_profit_std_dev_filter'] = 0

df_abc['dt_profit_both_filter'] = 0
df_abc['rf_profit_both_filter'] = 0
df_abc['gbm_profit_both_filter'] = 0
df_abc['lr_profit_both_filter'] = 0
df_abc['mlp_profit_both_filter'] = 0

for i in tqdm(range(2, last_n_matches+1)):
  X_train_average = train_average(X_train, i)
  X_test_average = test_average(X_test, i)
  X_train_scaled = scale(X_train_average)
  X_test_scaled = scale(X_test_average)
  gb, rf, dt, lr, mlp = run_models(X_train_scaled, y_train)
  
  y_pred_dt, y_pred_dt_proba, f1score_test_dt, y_pred_rf, y_pred_rf_proba, f1score_test_rf, y_pred_gb, y_pred_gb_proba, f1score_test_gb, y_pred_lr, y_pred_lr_proba, f1score_test_lr, y_pred_mlp, y_pred_mlp_proba, f1score_test_mlp = get_model_results(X_train_scaled, X_test_scaled, y_train, y_test)
  
  dt_f1score, rf_f1score, gbm_f1score, lr_f1score, mlp_f1score = get_model_metrics(y_pred_dt, f1score_test_dt, y_pred_rf, f1score_test_rf, y_pred_gb, f1score_test_gb, y_pred_lr, f1score_test_lr, y_pred_mlp, f1score_test_mlp, X_test_scaled, y_test)
  
  df_abc.dt.loc[i] = dt_f1score
  df_abc.rf.loc[i] = rf_f1score
  df_abc.gbm.loc[i] = gbm_f1score
  df_abc.lr.loc[i] = lr_f1score
  df_abc.mlp.loc[i] = mlp_f1score

  cnt = 1
  
  for k in [[y_pred_dt, y_pred_dt_proba], [y_pred_rf, y_pred_rf_proba], [y_pred_gb, y_pred_gb_proba], [y_pred_lr, y_pred_lr_proba], [y_pred_mlp, y_pred_mlp_proba]]:
    pred = pd.DataFrame(k[0])
    pred.columns = ['prediction']

    actual = pd.DataFrame(y_test)

    final_dff = pd.concat([X_test.reset_index(),pred.reset_index(), actual.reset_index()], axis = 1).drop('index', axis = 1)
    final_dff['bet'] = 100
    final_dff['roi'] = 0
    final_dff['profit'] = 0
    final_dff['vig'] = 0
    final_dff['max_prob'] = 0
    final_dff['std_dev'] = 0
    final_dff['-1 std dev'] = 0 
    final_dff['1 std dev'] = 0
    final_dff['std_dev_filter'] = 0

    
    #########################################################################
    final_dff['prob_diff_filter'] = False
    #########################################################################

    for j in range(0, final_dff.shape[0]):
      if final_dff['class_label'].iloc[j] == final_dff['prediction'].iloc[j]:

        odds_string = 'opening_odds_' + final_dff['class_label'].iloc[j]
        if final_dff[odds_string].iloc[j] >= 100:
          odds = final_dff[odds_string].iloc[j]/100 + 1
        elif final_dff[odds_string].iloc[j] <= -100:
          odds = 1- (100/final_dff[odds_string].iloc[j])
        else:
          odds = final_dff[odds_string].iloc[j]

      else:
        odds = 0

      roi = odds*100
      final_dff['roi'].iloc[j] = roi
      final_dff['profit'].iloc[j] = roi - 100


    ########################################################################

    for j in range(0, final_dff.shape[0]):
      if final_dff['opening_odds_home'].iloc[j] >= 100:
        final_dff['opening_odds_home'].iloc[j] = final_dff['opening_odds_home'].iloc[j]/100 + 1
      else:
        if final_dff['opening_odds_home'].iloc[j] <= -100:
          final_dff['opening_odds_home'].iloc[j] = 1-(100/final_dff['opening_odds_home'].iloc[j])
        else:
          final_dff['opening_odds_home'].iloc[j] = final_dff['opening_odds_home'].iloc[j]
      ####
      if final_dff['opening_odds_away'].iloc[j] >= 100:
        final_dff['opening_odds_away'].iloc[j] = final_dff['opening_odds_away'].iloc[j]/100 + 1
      else:
        if final_dff['opening_odds_away'].iloc[j] <= -100:
          final_dff['opening_odds_away'].iloc[j] = 1-(100/final_dff['opening_odds_away'].iloc[j])
        else:
          final_dff['opening_odds_away'].iloc[j] = final_dff['opening_odds_away'].iloc[j]
      #####
      if final_dff['opening_odds_draw'].iloc[j] >= 100:
        final_dff['opening_odds_draw'].iloc[j] = final_dff['opening_odds_draw'].iloc[j]/100 + 1
      else:
        if final_dff['opening_odds_draw'].iloc[j] <= -100:
          final_dff['opening_odds_draw'].iloc[j] = 1-(100/final_dff['opening_odds_draw'].iloc[j])
        else:
          final_dff['opening_odds_draw'].iloc[j] = final_dff['opening_odds_draw'].iloc[j]



      
  
    final_dff['opening_odds_home_prob'] = 1/final_dff['opening_odds_home']
    final_dff['opening_odds_away_prob'] = 1/final_dff['opening_odds_away']   
    final_dff['opening_odds_draw_prob'] = 1/final_dff['opening_odds_draw']

    for j in range(0, final_dff.shape[0]):    
      final_dff['vig'].iloc[j] = (final_dff['opening_odds_home_prob'].iloc[j]+final_dff['opening_odds_away_prob'].iloc[j]+final_dff['opening_odds_draw_prob'].iloc[j]) - 1

    final_dff['opening_odds_home_prob_wovig'] = final_dff['opening_odds_home_prob'] - final_dff['vig']/3
    final_dff['opening_odds_away_prob_wovig'] = final_dff['opening_odds_away_prob'] - final_dff['vig']/3
    final_dff['opening_odds_draw_prob_wovig'] = final_dff['opening_odds_draw_prob'] - final_dff['vig']/3

    pred_proba = k[1]
    pred_proba = pd.DataFrame(pred_proba)
    pred_proba.columns = ['pred_proba_away', 'pred_proba_draw', 'pred_proba_home']

    final_dff['pred_proba_home'] = pred_proba['pred_proba_home']
    final_dff['pred_proba_away'] = pred_proba['pred_proba_away']
    final_dff['pred_proba_draw'] = pred_proba['pred_proba_draw']



    ##############################################
    for j in range(0, final_dff.shape[0]):  
      if final_dff['prediction'].iloc[j]  == 'home':
        final_dff['max_prob'].iloc[j] = final_dff['pred_proba_home'].iloc[j]
        final_dff['prob_diff_filter'].iloc[j] = final_dff['pred_proba_home'].iloc[j]  >  final_dff['opening_odds_home_prob_wovig'].iloc[j]
      else:
        if final_dff['prediction'].iloc[j]  == 'away':
          final_dff['max_prob'].iloc[j] = final_dff['pred_proba_away'].iloc[j]
          final_dff['prob_diff_filter'].iloc[j] = final_dff['pred_proba_away'].iloc[j]  >  final_dff['opening_odds_away_prob_wovig'].iloc[j]
        else:
          final_dff['max_prob'].iloc[j] = final_dff['pred_proba_draw'].iloc[j]
          final_dff['prob_diff_filter'].iloc[j] = final_dff['pred_proba_draw'].iloc[j]  >  final_dff['opening_odds_draw_prob_wovig'].iloc[j]
  
    ###############################################

    final_dff['std_dev'] = np.std(final_dff['max_prob'])
    final_dff['mean'] = np.mean(final_dff['max_prob'])

    final_dff['-1 std dev'] = final_dff['mean'] - final_dff['std_dev']
    final_dff['1 std dev'] = final_dff['mean'] + final_dff['std_dev']

    final_dff['std_dev_filter'] = ((final_dff['max_prob'] > final_dff['-1 std dev'])&(final_dff['max_prob'] < final_dff['1 std dev']))




    profit_pct = final_dff['profit'].sum()*100/final_dff['bet'].sum()
    profit_pct_with_prob_diff_filter = final_dff[final_dff['prob_diff_filter'] == True]['profit'].sum()*100/final_dff[final_dff['prob_diff_filter'] == True]['bet'].sum()
    profit_pct_with_std_dev_filter = final_dff[final_dff['std_dev_filter'] == True]['profit'].sum()*100/final_dff[final_dff['std_dev_filter'] == True]['bet'].sum()
    profit_pct_with_both_filter = final_dff[(final_dff['prob_diff_filter'] == True)&(final_dff['std_dev_filter'] == True)]['profit'].sum()*100/final_dff[(final_dff['prob_diff_filter'] == True)&(final_dff['std_dev_filter'] == True)]['bet'].sum()
    
    if cnt == 1:
      df_abc.dt_profit.loc[i] = profit_pct
      df_abc.dt_profit_prob_diff_filter.loc[i] = profit_pct_with_prob_diff_filter
      df_abc.dt_profit_std_dev_filter.loc[i] = profit_pct_with_std_dev_filter
      df_abc.dt_profit_both_filter.loc[i] = profit_pct_with_both_filter
      final_dff_dt = final_dff
    elif cnt == 2:
      df_abc.rf_profit.loc[i] = profit_pct
      df_abc.rf_profit_prob_diff_filter.loc[i] = profit_pct_with_prob_diff_filter
      df_abc.rf_profit_std_dev_filter.loc[i] = profit_pct_with_std_dev_filter
      df_abc.rf_profit_both_filter.loc[i] = profit_pct_with_both_filter
      final_dff_rf = final_dff
    elif cnt == 3:
      df_abc.gbm_profit.loc[i] = profit_pct
      df_abc.gbm_profit_prob_diff_filter.loc[i] = profit_pct_with_prob_diff_filter
      df_abc.gbm_profit_std_dev_filter.loc[i] = profit_pct_with_std_dev_filter
      df_abc.gbm_profit_both_filter.loc[i] = profit_pct_with_both_filter
      final_dff_gbm = final_dff
    elif cnt == 4:
      df_abc.lr_profit.loc[i] = profit_pct
      df_abc.lr_profit_prob_diff_filter.loc[i] = profit_pct_with_prob_diff_filter
      df_abc.lr_profit_std_dev_filter.loc[i] = profit_pct_with_std_dev_filter
      df_abc.lr_profit_both_filter.loc[i] = profit_pct_with_both_filter
      final_dff_lr = final_dff
    elif cnt == 5:
      df_abc.mlp_profit.loc[i] = profit_pct
      df_abc.mlp_profit_prob_diff_filter.loc[i] = profit_pct_with_prob_diff_filter
      df_abc.mlp_profit_std_dev_filter.loc[i] = profit_pct_with_std_dev_filter
      df_abc.mlp_profit_both_filter.loc[i] = profit_pct_with_both_filter
      final_dff_mlp = final_dff
    
    cnt = cnt+1


profit_pct = final_dff['profit'].sum()*100/final_dff['bet'].sum()
df_abc.to_excel('/drive/My Drive/Automation Script/Modeling output/'+league_name+'.xlsx')
print('File written out to drive!')
```
